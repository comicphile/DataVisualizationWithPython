```python
#Basic operations with dataframes

#multiplication
result=stats.BirthRate * stats.InternetUsers
print(result.head())
print('--------------------------------------------------------------------------------------------------')

#Add new column
stats['newCol']=result
print(stats.head())
print('---------------------------------------------------------------------------------------------------')

#Deleting a column
stats=stats.drop('newCol',axis=1)
print(stats.head())
```

---

```python
#Filtering a dataframe
Filter=stats.InternetUsers<2  #An object that stores the boolen values, for every row
print(Filter.head())
print('--------------------------------------------------------------------------------------------')
print(stats[Filter])          #Prints the records meeting the condition stored in boolean, in this case Internet Users < 2
print('--------------------------------------------------------------------------------------------')
print(stats[stats.BirthRate>40])   #Prints the rocords with Birth rate > 40/10000
print('---------------------------------------------------------------------------------------------')
#print(stats[Filter and stats.BirthRate>40])  generates error because of ambiguity of comparison of both series of boolean filters
Filter2=stats.BirthRate>40
print(stats[Filter & Filter2])
print()
print(stats[(stats.InternetUsers<2) & (stats.BirthRate>40)]) #Using combination of filters
print('---------------------------------------------------------------------------------------------')
print(stats[stats.IncomeGroup=='High income'])
```

```python
#using .at() and .iat() for accessing individual elements
#stats[2][2]   generates error
print(stats.head(4))
print('-------------------------------------------------------------------')
print(stats.iat[2,2])            #Accesses element using integral location, column and row heads are not counted
print('-------------------------------------------------------------------')
print(stats.at[2,'BirthRate'])   #Accesses element using their respective row label and column label
```

```python
#demonstrating difference in working of .iat and .at
subs10=stats[::10]  #subset with positive 10 increment
print(subs10)
print('--------------------------------------------------------------------')
print(subs10.iat[10,0])              #accesses at row index 10 or 11th row and column index 0 or Country Name
print('--------------------------------------------------------------------')
print(subs10.at[10,'CountryName'])   #accesses at row label 10 and coumn- Country Name
```
