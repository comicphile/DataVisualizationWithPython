```python
#Introduction to Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import os
plt.rcParams['figure.figsize']=8,4   #to increase size of the plot
#print(os.getcwd())

stats=pd.read_csv("C:\\Users\\Madhav\\Desktop\\PAZ Code\\Dataset\\DemographicData.csv")
stats.columns=['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers','IncomeGroup']
#stats

vis1=sns.distplot(stats["InternetUsers"])    #distplot plots visualization using Seaborn
```

```python
vis1=sns.distplot(stats['InternetUsers'],bins=30)#a hump between 0 and 20 suggests quite a few countries have 20% internet access
```

```python
#boxplot
vis2=sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate')  #visualizes incomegroup vs birthrate, eg- high income group has low
                                                             #birthrate
```

```python
#lmplot- linear regression model
vis3=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False)  #fit_reg to remove linear regression plotline
#high birth rate and low no. of internet users is co-related here
```

```python
vis4=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',size=8,aspect=1) #bringing income group in mix
```

```python
#Keyword arguments in Python
vis5=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',size=8,aspect=1,\
               scatter_kws={'s':100})  #using scatter_kws dictionary to increase the size of the marker by sending the parameter
                                    #to plt.scatter() around which sns.lmplot is wrapped,in which some of the former's parameters
                                    #aren't available
```
